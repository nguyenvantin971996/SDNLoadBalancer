```python
import os
import random
import time

from BMA import BMA
from ABC import ABC
from GA import GA
from Draw_3D_chart import BarChart

N = 20
Max = 100
K_paths = 10
w = 2
c1 = 1
c2 = 0.1
Pc = 0.9
Pm = 0.9
Ts = 2
dst = 2
src = 7
alg_1 = BMA(dst,src,N, Max, K_paths, w, c1, c2)
alg_1.Do()
alg_2 = ABC(dst,src,N, Max, K_paths)
alg_2.Do()
alg_3 = GA(dst,src,N, Max, K_paths, Pc, Pm, Ts)
alg_3.Do()
pr = ["BMA","ABC","GA"]
pr.append("Type of algorithm")
chart = BarChart(alg_1.values,alg_2.values,alg_3.values,"BMA_ABC_GA",pr)
chart.Do()
f=open("../../demo.txt","w")
f.truncate(0)
for i in range(len(alg_1.best)):
    stt = ",".join(str(x) for x in alg_1.best[i].path)
    stt= stt+","+str(alg_1.best[i].fitness)+"\n"
    f.write(stt)
f.close()
```
